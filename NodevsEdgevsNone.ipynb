```python
import matplotlib.pyplot as plt
import proplot as plot
import numpy as np
import pandas as pd
plot.rc['title.loc'] = 'uc'
plot.rc['axes.ymargin'] = plot.rc['axes.xmargin'] = 0.05
plot.rc.small=16
plot.rc.large=24
```

```python
import pickle as pkl
```

```python
with open("citeseer.pkl",'rb') as f:
  citeseer = pkl.load(f)
with open("cora.pkl",'rb') as f:
  cora = pkl.load(f)
with open("human.pkl",'rb') as f:
  human = pkl.load(f)
with open("pubmed.pkl",'rb') as f:
  pubmed = pkl.load(f)
with open("wordnet.pkl",'rb') as f:
  wordnet = pkl.load(f)
with open("yeast.pkl",'rb') as f:
  yeast = pkl.load(f)
```

```python
citeseer_ds = np.array([citeseer['none'], citeseer['node'], citeseer['edge']]).T
cora_ds = np.array([cora['none'], cora['node'], cora['edge']]).T
human_ds = np.array([human['none'], human['node'], human['edge']]).T
pubmed_ds = np.array([pubmed['none'], pubmed['node'], pubmed['edge']]).T
wordnet_ds = np.array([wordnet['none'], wordnet['node'], wordnet['edge']]).T
yeast_ds = np.array([yeast['none'], yeast['node'], yeast['edge']]).T
```

```python
citeseer_ds.shape
```

```python
f, axs = plot.subplots(
    nrows=1, ncols=6, aspect=1.5, axwidth=4,
    share=0, wratios=(1,1,1,1,1,1)
)

ax1 = axs[0]
ax1.boxplot(citeseer_ds, showfliers=False, labels=['None', 'Node', 'Edge'])
ax1.set_title('Citeseer')

ax2 = axs[1]
ax2.boxplot(cora_ds, showfliers=False, labels=['None', 'Node', 'Edge'])
ax2.set_title('Cora')

ax3 = axs[2]
ax3.boxplot(pubmed_ds, showfliers=False, labels=['None', 'Node', 'Edge'])
ax3.set_title('Pubmed')

ax4 = axs[3]
ax4.boxplot(human_ds, showfliers=False, labels=['None', 'Node', 'Edge'])
ax4.set_title('Human')

ax4 = axs[4]
ax4.boxplot(wordnet_ds, showfliers=False, labels=['None', 'Node', 'Edge'])
ax4.set_title('Wordnet')

ax5 = axs[5]
ax5.boxplot(yeast_ds, showfliers=False, labels=['None', 'Node', 'Edge'])
ax5.set_title('Yeast')
```
