```python
import numpy as np
import pandas as pd
```

```python
yeast= []
for tout in [0.1, 0.2, 0.5, 1]:
    for t in ['iso', 'lru', 'recache', 'ours']:
        for dataset in ['yeast']:        
            abc = np.load("exp5/"+dataset+"_"+str(tout*10)+"_" + t + ".npy")
            abc = abc/100
            yeast.append(abc)
y = pd.DataFrame(yeast, index=[[0.1,0.1,0.1,0.1,0.2,0.2,0.2,0.2,0.5,0.5,0.5,0.5,1,1,1,1],['iso', 'lru', 'recache', 'ours','iso', 'lru', 'recache', 'ours','iso', 'lru', 'recache', 'ours','iso', 'lru', 'recache', 'ours']])
```

```python
human= []
for tout in [0.1, 0.2, 0.5, 1]:
    for t in ['iso', 'lru', 'recache', 'ours']:
        for dataset in ['human']:        
            human.append(np.load("exp5/"+dataset+"_"+str(tout*10)+"_" + t + ".npy")/100)
h = pd.DataFrame(human, index=[[0.1,0.1,0.1,0.1,0.2,0.2,0.2,0.2,0.5,0.5,0.5,0.5,1,1,1,1],['iso', 'lru', 'recache', 'ours','iso', 'lru', 'recache', 'ours','iso', 'lru', 'recache', 'ours','iso', 'lru', 'recache', 'ours']])
```
